```python
assets = {
    "Brent": "BZ=F",
    "WTI": "CL=F",
    "Gold": "GC=F",
    "SP500": "^GSPC",
    "NASDAQ": "^IXIC",
    "VIX": "^VIX",
    "USD_Index": "DX-Y.NYB",
    "INR_USD": "INR=X",
    
    "Lockheed": "LMT",
    "Raytheon": "RTX",
    "Delta": "DAL",
    "American_Airlines": "AAL"
}
```

```python
market_data = {}
```

```python
for name, ticker in assets.items():
    
    data = yf.download(
        ticker,
        start="2020-01-01",
        end="2026-04-30"
    )
    
    data.columns = ['Close', 'High', 'Low', 'Open', 'Volume']
    
    data.reset_index(inplace=True)
    
    market_data[name] = data
    
    print(f"{name} data downloaded")
```

```python
market_data['Gold'].head()
```

```python
market_data["SP500"][["Date", "Close"]].head()
```

```python
brent_df = market_data["Brent"][["Date", "Close"]].copy()

gold_df = market_data["Gold"][["Date", "Close"]].copy()
```

```python
brent_df.rename(columns={"Close": "Brent"}, inplace=True)

gold_df.rename(columns={"Close": "Gold"}, inplace=True)
```

```python
merged_df = pd.merge(
    brent_df,
    gold_df,
    on="Date"
)

merged_df.head()
```

```python
master_df = None

for name, data in market_data.items():
    
    temp_df = data[["Date", "Close"]].copy()
    
    temp_df.rename(columns={"Close": name}, inplace=True)
    
    if master_df is None:
        master_df = temp_df
    else:
        master_df = pd.merge(
            master_df,
            temp_df,
            on="Date"
        )

master_df.head()
```

```python
master_df.isnull().sum()
```

```python
master_df.to_csv("master_market_data.csv", index=False)
```

```python
import yfinance as yf
import pandas as pd
```

```python
natural_gas = yf.download(
    "NG=F",
    start="2020-01-02",
    end="2026-05-11"
)

wheat = yf.download(
    "ZW=F",
    start="2020-01-02",
    end="2026-05-11"
)

copper = yf.download(
    "HG=F",
    start="2020-01-02",
    end="2026-05-11"
)
```

```python
natural_gas = natural_gas[["Close"]]
natural_gas.columns = ["NaturalGas"]

wheat = wheat[["Close"]]
wheat.columns = ["Wheat"]

copper = copper[["Close"]]
copper.columns = ["Copper"]
```

```python
natural_gas = natural_gas.reset_index()
wheat = wheat.reset_index()
copper = copper.reset_index()
```

```python
master_df["Date"] = pd.to_datetime(master_df["Date"])

natural_gas["Date"] = pd.to_datetime(natural_gas["Date"])
wheat["Date"] = pd.to_datetime(wheat["Date"])
copper["Date"] = pd.to_datetime(copper["Date"])
```

```python
master_df = master_df.merge(
    natural_gas,
    on="Date",
    how="left"
)

master_df = master_df.merge(
    wheat,
    on="Date",
    how="left"
)

master_df = master_df.merge(
    copper,
    on="Date",
    how="left"
)
```

```python
master_df[
[
"Date",
"NaturalGas",
"Wheat",
"Copper"
]
].head()
```

```python
master_df["NaturalGas_Return"] = (
    master_df["NaturalGas"].pct_change() * 100
)

master_df["Wheat_Return"] = (
    master_df["Wheat"].pct_change() * 100
)

master_df["Copper_Return"] = (
    master_df["Copper"].pct_change() * 100
)
```

```python
master_df[
[
"Date",
"NaturalGas_Return",
"Wheat_Return",
"Copper_Return"
]
].head()
```

```python
master_df.to_csv(
    "master_market_features.csv",
    index=False
)
```
